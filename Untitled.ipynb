```python
import requests
import pandas as pd

API_KEY = '737b518969651f0e3c9a540e'
url = f'https://v6.exchangerate-api.com/v6/737b518969651f0e3c9a540e/latest/USD'

response = requests.get(url)
data = response.json()

currencies = ['EGP', 'SAR', 'AED', 'GBP', 'JPY', 'EUR', 'CAD']
rates = {currency: data['conversion_rates'][currency] for currency in currencies}


df = pd.DataFrame(rates.items(), columns=['Currency', 'Exchange Rate'])
df['Base Currency'] = data['base_code']
df['Date'] = data['time_last_update_utc']

print(df)
```

```python
df.to_csv("exchange_rates.csv", index=False)
print("✅ Saved as exchange_rates.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df['Currency'], df['Exchange Rate'], color='teal')
plt.title(f"Exchange Rates vs USD\n({df['Date'][0]})")
plt.xlabel("Currency")
plt.ylabel("Rate")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
